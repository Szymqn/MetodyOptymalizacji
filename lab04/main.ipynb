```python
import math
from package_.plots import plot_one_variable_with_range
```

```python
A = 1
B = 6 
EPSILON = 0.001
K = (math.sqrt(5) - 1) / 2
```

```python
f = lambda x: 2 * x ** 3 + 3 * x ** 2 - 48 * x + 13 
x_1, x_2 = B - K * (B - A), A + K * (B - A)
```

```python
def max(x_1=x_1, x_2=x_2, k=K, a=A, b=B, f=f):
    while abs(x_2 - x_1) > EPSILON:
        if f(x_1) > f(x_2):
            b = x_2
            x_2 = x_1
            x_1 = b - k * (b - a)
        else:
            a = x_1
            x_1 = x_2
            x_2 = a + k * (b - a)
            
        plot_one_variable_with_range(A, B, x_1, x_2, f)
    
    return (a + b) / 2
```

```python
def min(x_1=x_1, x_2=x_2, k=K, a=A, b=B, f=f):
    while abs(x_2 - x_1) > EPSILON:
        if f(x_1) < f(x_2):
            b = x_2
            x_2 = x_1
            x_1 = b - k * (b - a)
        else:
            a = x_1
            x_1 = x_2
            x_2 = a + k * (b - a)
            
        plot_one_variable_with_range(A, B, x_1, x_2, f)
    
    return (a + b) / 2
```
